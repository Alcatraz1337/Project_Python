```python
import math
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
def sin_taylor(x, n):
    y = []
    for j in x:
        sum = 0
        for i in range(0, n + 1):
            sum += (-1)**i / math.factorial(2 * i + 1) * j**(2 * i + 1)
        y.append(sum)
    return y
```

```python
def exp_taylor(x, n):
    y = []
    for j in x:
        sum = 0
        for i in range(0, n + 1):
            sum += j**i / math.factorial(i)
        y.append(sum)
    return y
```

```python
x = np.arange(-5, 5, 0.1)
y_taylor_1 = sin_taylor(x, 1)
y_taylor_5 = sin_taylor(x, 5)
y_taylor_10 = sin_taylor(x, 10)
y_taylor_20 = sin_taylor(x, 20)
y_sin = []
for i in x:
    y_sin.append(np.sin(i))
```

```python
plt.figure(1)
ax = plt.gca()
ax.set_aspect(1)
ax.set_ylim([-5,5])
plt.plot(x, y_taylor_1, color='r', label = 'N = 1')
plt.plot(x, y_taylor_5, c = 'g', label = 'N = 5')
plt.plot(x, y_taylor_10, c = 'y', label = 'N = 10')
plt.plot(x, y_taylor_20, c = 'orange', label = 'N = 20')
plt.plot(x, y_sin, color = 'b', label = 'Sin(x)')
plt.legend()
plt.show()
```

```python
sin_deviation = []
sin_deviation.append(np.sum(np.abs(np.array(y_sin) - np.array(y_taylor_1))))
sin_deviation.append(np.sum(np.abs(np.array(y_sin) - np.array(y_taylor_5))))
sin_deviation.append(np.sum(np.abs(np.array(y_sin) - np.array(y_taylor_10))))
sin_deviation.append(np.sum(np.abs(np.array(y_sin) - np.array(y_taylor_20))))

x_label = ['N = 1', 'N = 5', 'N = 10', 'N = 20']
txt_xPos = [0, 1, 2, 3]
plt.title('Deviation')
plt.bar(x_label, sin_deviation)
```

```python
plt.figure(1)
ax = plt.gca()
ax.set_aspect(1)
ax.set_ylim([-5,5])
y_taylor_1 = exp_taylor(x, 1)
y_taylor_5 = exp_taylor(x, 5)
y_taylor_10 = exp_taylor(x, 10)
y_taylor_20 = exp_taylor(x, 20)
y_exp = []
for i in x:
    y_exp.append(np.exp(i))

plt.plot(x, y_taylor_1, color='r', label = 'N = 1')
plt.plot(x, y_taylor_5, c = 'g', label = 'N = 5')
plt.plot(x, y_taylor_10, c = 'y', label = 'N = 10')
plt.plot(x, y_taylor_20, c = 'orange', label = 'N = 20')
plt.plot(x, y_exp, color = 'b', label = 'Exp(x)')
plt.legend()
plt.show()
```

```python
exp_deviation = []
exp_deviation.append(np.sum(np.abs(np.array(y_exp) - np.array(y_taylor_1))))
exp_deviation.append(np.sum(np.abs(np.array(y_exp) - np.array(y_taylor_5))))
exp_deviation.append(np.sum(np.abs(np.array(y_exp) - np.array(y_taylor_10))))
exp_deviation.append(np.sum(np.abs(np.array(y_exp) - np.array(y_taylor_20))))

x_label = ['N = 1', 'N = 5', 'N = 10', 'N = 20']
txt_xPos = [0, 1, 2, 3]
plt.title('Deviation')
plt.bar(x_label, exp_deviation)
```
