```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
```

# 課題２

```python
sampleNUm = 10000
mu, sigma = 0, 2
```

```python
rands = np.random.normal(mu, sigma, sampleNUm)
y = st.norm.pdf(rands, mu, sigma) 
avg = np.mean(rands)
stdDev = np.std(rands)
```

```python
print("Average:", avg, " Standard Deviation:", stdDev)
```

```python
zipped = zip(rands, y)
rands, y = zip(*sorted(zipped)) # Sort x, y
```

```python
plt.hist(rands, bins=40, density= True) # Histogram
plt.plot(rands, y, 'r--') # Plot 
plt.show()
```

# 課題３

### More samples

```python
SampleNum_1 = 100000
sum = 0
inRange = []
outRange = []
```

```python
for i in range(0,SampleNum_1):
    x = np.random.rand(1)
    y = np.random.rand(1)

    if (np.square(x) + np.square(y)) < 1:
        sum += 1
        inRange.append(tuple((x, y)))
    else:
        outRange.append(tuple((x, y)))
```

```python
plt.axis('equal')
plt.scatter(*zip(*inRange), s = 1, c = 'r')
plt.scatter(*zip(*outRange), s = 1, c = 'b')
P = sum * 4 / SampleNum_1
plt.show()
print("Pi as calculated:", P)
```
