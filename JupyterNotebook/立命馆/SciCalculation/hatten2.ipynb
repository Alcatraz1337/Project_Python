```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
```

```python
def Z():
    return 1 / 2 - 1 / 3
    
```

```python
def p(x):
    return x * (1 - x) / Z()
```

```python
def q(x):
    return 1
```

```python
def getSample():
    return np.random.rand()
```

```python
sampleNum = 100000
k = 1.5
accepted = []
for i in range(sampleNum):
    accept = False
    while not accept:
        x = getSample()
        sample = np.random.uniform(0, k * q(x))
        if sample < p(x):
            accept = True
            accepted.append(x)
```

```python
accepted = sorted(accepted)
plt.hist(accepted, bins=40, weights=np.ones(len(accepted)) / len(accepted))
plt.show()
```
