<h3> 1．　総合特性値</h3><br>

標準化データ$y_{np}$はサンプルデータ$x_{np}(n=1, \cdots, N;p \in \{1,\cdots,P\})$の平均$\mu_p$および分散$\sigma_p^2$より平均＝0、分散＝1となるように変形したものである。<br>

\begin{align}
\mu_p &= \frac{1}{N}\sum_{n=1}^{N}x_{np}\\
\sigma_p^2 &= \frac{1}{N}\sum_{n=1}^{N}(x_{np}-\mu_p)^2\\
y_{np} &= \frac{x_{np}-\mu_p}{\sigma_p}
\end{align}

標準化データ$y_{np}$を用いて$L_{mp}$を重み係数とした総合特性値（合成変数）$z_{nm}(m=1, \cdots, M; M < P)$を考える。ここで、$z_{nm}$はサンプル$n$における第$m$主成分という。<br>

\begin{equation}
	\left.
	\begin{array}{ccc}
  	z_{n1} & = & L_{11}y_{n1}+L_{12}y_{n2}+ \dots +L_{1P}y_{nP} \\
  	z_{n2} & = & L_{21}y_{n1}+L_{22}y_{n2}+ \dots +L_{2P}y_{nP} \\  
  	    & \dots &  \\
  	z_{nM} & = & L_{M1}y_{n1}+L_{M2}y_{n2}+ \dots +L_{MP}y_{nP} 
	\end{array}
	\right\}
\end{equation}

ただし、条件として

\begin{align}
L_{m1}^2 + L_{m2}^2 + \dots + L_{mP}^2 &= \sum_{p=1}^P L_{mp}^2\\
& \equiv 1
\end{align}

ここで、$L_{mp}$は上記条件の下で$z_{nm}(n=1, \cdots, N; m\in\{1, \cdots, M\})$の分散$\sigma_m^2$が最大となるように決定される。また、同時に以下の2点を満足しなければならない。

<ol>
    <li>$\sigma_1^2 \ge \sigma_2^2 \ge \dots \ge \sigma_M^2$</li>
    <li>
        $z_{nm}$と$z_{nm^\prime}$は□、ただし$(m \ne m^\prime; m, m^\prime \in \{1, \cdots, M\})$
    </li>
</ol>

<h3> 2．　第1主成分の導出</h3>

第1主成分$z_{n1}$は$M$個の主成分$z_{nm}(m=1, \cdots, M)$の中でその分散がもっとも大きい$\sigma_1^2$を持つ。そのときの重み係数$L_{1p}(p=1, \cdots, P)$が求まれば、第1主成分$z_{n1}$を求めることができる。

\begin{equation}
\sigma_1^2 = \max_m \sigma_m^2
\end{equation}


<h4>2.1 分散$\sigma_1^2$の計算</h4>

第1主成分$z_{n1}(n=1, \cdots, N)$の分散$\sigma_1^2$は

\begin{align}
\sigma_1^2 
&=\sum_{p=1}^{P}\sum_{p^\prime=1}^{P}L_{1p}L_{1p^\prime}\frac{1}{N}\sum_{n=1}^{N}y_{np}y_{np'}
\end{align}

一方、$y_{np}$と$y_{np^\prime}$の相関係数$r_{pp^\prime}(p, p^\prime \in \{1,2, \cdots, P\})$は

\begin{align}
r_{pp^\prime}
&=\frac{1}{N}\sum_{n=1}^{N}y_{np}y_{np^\prime}
\end{align}

このように要素$r_{pp^\prime}$を有する$P \times P$行列を相関行列${R}$という。

\begin{equation}
R = \left( r_{pp^\prime} \right) = 
	\left(
    \begin{array}{cccc}
      r_{11} & r_{12} & \ldots & r_{1P} \\
      r_{21} & r_{22} & \ldots & r_{2P} \\
      \vdots & \vdots & \ddots & \vdots \\
      r_{P1} & r_{P2} & \ldots & r_{PP}
    \end{array}
  \right)
\end{equation}

以上より

\begin{equation}
\sigma_1^2 = \sum_{p=1}^{P}\sum_{p^\prime=1}^{P}L_{1p}L_{1p^\prime}r_{pp^\prime}
\end{equation}


<h4>2.2 分散$\sigma_1^2$の最大化</h4>

一般的に制約条件付き最大化問題はLagrangeの未定乗数法によって解くことができる。いま、先の制約条件の式から

\begin{align}
\sum_{p=1}^P L_{1p}^2　& =1 
\end{align}

Lagrange関数$F(L_{11}, \cdots, L_{1P}, \lambda)$の最大化を考える。

\begin{equation}
F = \sum_{p=1}^{P}\sum_{p^\prime=1}^{P}L_{1p}L_{1p^\prime}r_{pp^\prime}-\lambda(\sum_{p=1}^{P}L_{1p}^2-1)
\end{equation}

$F$を$L_{1p}(p=1, \cdots, P)$と$\lambda$でそれぞれ偏微分して0とおき、$P$個の$p$元1次連立方程式を解くことにより$L_{1p}(p=1, \cdots, P)$を求めることができる。

\begin{equation}
\sum_{p^\prime=1}^P L_{1p^\prime} r_{pp^\prime} - \lambda L_{1p} = 0 
\end{equation}

<h4>2.3固有値の計算</h4>

上記の式は以下のような行列式に書き直すことができる。

\begin{equation}
	\left(
    \begin{array}{cccc}
      r_{11}-\lambda & r_{12} & \ldots & r_{1P} \\
      r_{21} & r_{22}-\lambda & \ldots & r_{2P} \\
      \vdots & \vdots & \ddots & \vdots \\
      r_{P1} & r_{P2} & \ldots & r_{PP}-\lambda
    \end{array}
  \right)
\left(
	\begin{array}{c}
	L_{11} \\
	L_{12} \\
	\vdots\\
	L_{1P} \\
	\end{array}
\right)
= 0
\end{equation}

これは次のように表現できる。ただし、$I$は単位行列である。

\begin{equation}
(R-\lambda I)L_1 = 0
\end{equation}

上式は固有値問題に帰着できるので、固有方程式$□$を解けばよいことになる。得られた解$\lambda_p(p=1, \cdots, P)$は相関行列$R$の□、$L_{1p}$は$R$の□と呼ばれる。

ここで、最大の$\lambda_p$を$\lambda_1$とし、この$\lambda_1$と$R$の相関係数$r_{pp^\prime}$より、具体的な$L_{1p}(p=1, \cdots, P)$の値を求めることができる。

また、第1主成分$z_{n1}$の分散は相関行列$R$の最大固有値となる。

\begin{align}
\sigma_1^2 &= \sum_{p=1}^P \sum_{p^\prime=1}^P L_{1p}L_{1p^\prime} r_{pp^\prime} \\
&= \lambda_1 
\end{align}

<h3> 3．　主成分数$M$の決定および寄与率の計算</h3>

主成分数$M$の決定については、以下のような方法が提案されている。
<ul>
    <li> 相関行列$R$の固有値の中で□より大きい固有値の数</li>
    <li> 固有値が大きく下がる前までの主成分数（スクリープロット法）</li>
</ul>

$P$個の変数に対して$P$個の主成分を求めることができる。寄与率（contribution ratio）とは、第$m(m=1, \cdots, P)$主成分は元のデータをどの程度説明しているかの尺度である。また、寄与率を第1主成分から順に累積していったものを累積寄与率（cumulative contribution ratio）と呼ぶ。第$m$主成分寄与率$cr_m$およ
び第$m$主成分までの累積寄与率$ccr_m$は以下のように定義される。

\begin{align}
cr_m &= \frac{\lambda_{m}}{P}\\
ccr_m &= \frac{\sum_{i=1}^m \lambda_i}{P}
\end{align}

<h3> 4．　主成分得点</h3>

主成分得点（principal component scores）とは、主成分数$M$、重み係数$L_{mp}$を決定したのち、サンプル$n(n=1, \cdots, N)$の第$m(m=1, \cdots, M)$主成分$z_{nm}$に標準化データ$y_{np}$の値を代入して求めた数値であり、サンプル$n$の主成分$m$軸上の値を示す。第$m$主成分得点の分散は$\lambda_m$である。また、第$m$主成分得点の平均は0となる。

<h3> 5．　主成分負荷量</h3>

主成分負荷量（principal component loadings）とは、標準化データ$y_{np}$と主成分$z_{nm}$の相関係数$r_{pm}(p \in \{1,\cdots,P\}; m \in \{1,\cdots,M\})$であり、胃kのように計算できる。ただし、$\mathrm{cov}(y_{np},z_{nm})$は$y_{np}$と$z_{nm}$の分散共分散行列を表す。

\begin{align}
r_{pm} &= \frac{\mathrm{cov}(y_{np},z_{nm})}{\sigma_p \sigma_m} \\
&= \sqrt{\lambda_{m}}L_{mp}
\end{align}

主成分負荷量は変数$p$と主成分$m$の関係を解釈する上で有用である。主成分$m$における主成分負荷量の二乗和は式固有値$\lambda_m$に等しい。

\begin{align}
\sum_{p=1}^P r_{pm}^2 &= \lambda_m
\end{align}

<h3> 7．　Pythonを用いた主成分分析</h3>

1. 1. manaba+Rより下記のサンプルデータファイルを取得せよ。
<ul>
        <li>F1クラス：data_pca_F1.csv</li>

2. Pythonでコーディングをして、サンプルデータの主成分分析を実行せよ。

```python
import pandas as pd
import numpy as np

COLUMN=2
data=pd.read_csv("data_pcs_f2.csv", sep=",")
print(data.head())
print("\n")

from scipy.stats import zscore
data_std=zscore(data)
print("standardized data=\n",data_std)
print("\n")

from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(data_std)

mn = np.mean(data_std,axis=0)
z = data_std - mn
cv = np.cov(z[:,0],z[:,1],bias=1)  #標本分散 (sample variance)

#cv=pca.get_covariance()  不偏分散 (unbiased variance) 

print ("covariance=\n",cv)
print("\n")

W, v = np.linalg.eig(cv)

W=np.sort(W)[::-1]
print("eigenvalue=\n",W)
print("\n")

print("eigenvector=\n",v) #固有ベクトルは列ベクトルになることに注意
print("\n")

print("n_components=",pca.n_components_)
print("\n")

print ("components=\n",pca.components_)
print("\n")

print("explained variance ratio\n",pca.explained_variance_ratio_)
print("\n")

score = pca.transform(data_std)
print ("scores=\n",score)
print("\n")

load=np.zeros((pca.n_components_,COLUMN))
for i in range(pca.n_components_):
    for j in range(COLUMN):
        load[i][j]=np.sqrt(W[i])*pca.components_[i][j]
print("loads=\n",load)

from matplotlib import pyplot as plt
transformed = pca.fit_transform(data_std)

plt.scatter(transformed[:, 0], transformed[:, 1])
plt.title('principal component')
plt.xlabel('pc1')
plt.ylabel('pc2')

plt.show()
```

3. 分散共分散行列（covariance）が相関行列$R$と一致することを確認せよ。 一致です
4. 固有値（eigenvalue）および固有ベクトル（eigenvector）がエクセルによる計算結果と一致することを確認せよ。　一致です
5. 主成分（components）が固有ベクトル（eigenvector）と一致していることを確認せよ。　一致です
6. 主成分得点（scores）および主成分負荷量（loads）がエクセルによる計算結果と一致することを確認せよ。　一致です
