<h3> 1．　基本的考え方</h3><br>

いま、群$A$に属する$N_A$個のサンプル$(x_{i1}^A, x_{i2}^A)(i=1,\cdots,N_A)$と群$B$に属する$N_B$個のサンプル$(x_{i1}^B, x_{i2}^B)(i=1,\cdots,N_B)$からなる$N$個（$N=N_A+N_B$）の2変量サンプル$(x_{1}, x_{2})$が次式で与えられる分割直線で最適に分割されるとする。

\begin{equation}
w _{1}x_{1}+w _{2}x_{2}+w _{0} =0
\end{equation}

そのとき、左辺を$z$に等置した方程式（関数）を線形判別関数$z$と呼ぶ。また、サンプル$(x_{i1}, x_{i2})(i \in N)$における$z$の値$z_i$を判別得点という。したがって、判別得点$z_i$の正負でサンプル$(x_{i1}, x_{i2})$が群$A$に属するか、群$B$に属するかを決定することができる。

\begin{align}
& z = w_1x_1 + w_2x_2 + w_0 \\
& z_i = w _{1}x_{i1}+w _{2}x_{i2}+w _{0}
\end{align}

サンプル$(x_{i1}, x_{i2})$と分割直線までの距離$d_i$は

\begin{align}
d_i &= \frac{| z_i |}{\sqrt{w_1^2 + w_2^2}}
\end{align}

すなわち、判別得点$z_i$は分割直線とサンプルとの離れ具合を表している。


<h3> 2．　フィッシャーの線形判別分析</h3>

いま、合成変量$f$を次のように定義する。

\begin{align}
f &= w^\mathrm{T}x\\
&= w_1x_1+w_2x_2
\end{align}

ここで、$w=(w_1,w_2)^\mathrm{T}$は線形変換ベクトル、$x=(x_1,x_2)^\mathrm{T}$はサンプルベクトルを表し、$f$は線形変換された1次元の値となる。ただし、全サンプルの平均を$\bar{x}=(\bar{x}_1,\bar{x}_2)^\mathrm{T}$と表したとき、$w^\mathrm{T}$は次式を満足することとする。

\begin{align}
& w^\mathrm{T}\bar{x} =-w_{0}\\
& \bar{x}_1 = \frac{1}{N}\sum_{i=1}^N x_{i1} \\
& \bar{x}_2 = \frac{1}{N}\sum_{i=1}^N x_{i2} 
\end{align}

すなわち、直線$f=w^\mathrm{T}x=-w_0$は$\bar{x}$を通り、$w$に垂直な直線$w_{1}x_{1}+w_{2}x_{2}+w_{0}=0$を意味している。

次に、合成変数$f$の全変動（偏差平方和）$S_T$を考える。

\begin{align}
S_T &= \sum_{i=1}^{N}(f_i-\bar{f})^2\\
&= S_B + S_W 
\end{align}

とおけば、

\begin{align}
S_B &= N_{A}\left ( \bar{z} ^{A}-\bar{z} \right )^{2}+N_{B}\left ( \bar{z} ^{B}-\bar{z} \right )^{2} \\
&= w^\mathrm{T}Bw\\
S_W &=\sum_{i\in A}^{N_{A}}\left ( z_{i}-\bar{z} ^{A}\right )^{2}+\sum_{i\in B}^{N_{B}}\left ( z _{i}- \bar{z} ^{B} \right )^{2}\\
&= w^\mathrm{T}Ww
\end{align}

$S_B$は$AとB$がどのくらい離れているかを表す量（between）、$S_W$は各群内での分散を表す量（within）と呼ばれる。<br>
ただし、

\begin{align}
& \bar{f} = -w_0\\
& \bar{f}^A = \frac{1}{N_A}\sum_{i\in A}^{N_A}(w_1x_{i1}^A+w_2x_{i2}^A) \\
& \bar{f}^B = \frac{1}{N_B}\sum_{i\in B}^{N_B}(w_1x_{i1}^B+w_2x_{i2}^B)\\
& B = \frac{N_AN_B}{N}\left( \bar{x}^A-\bar{x}^B\right)\left(\bar{x}^A-\bar{x}^B\right)^\mathrm{T}\\
& W = \left(
\begin{array}{cc}
S_{11}^A+S_{11}^B  & S_{12}^A+S_{12}^B \\
S_{12}^A+S_{12}^B & S_{22}^A+S_{22}^B
\end{array}
\right)
\end{align}

$S_{11}^A$：群$A$における$x_1$の偏差平方和<br>
$S_{11}^B$：群$B$における$x_1$の偏差平方和<br>
$S_{12}^A$：群$A$における$x_1とx_2$の偏差積和<br>
$S_{12}^B$：群$B$における$x_1とx_2$の偏差積和<br>
$S_{22}^A$：群$A$における$x_2$の偏差平方和<br>
$S_{22}^B$：群$B$における$x_2$の偏差平方和。

判別分析の目的は、群$A,B$を最も適切に分割する分割直線を見つけることである。したがって、群間変動である$S_B$が最大となる分割直線を探せばよい。そのために、次のような相関比$\eta^2$を定義し、それを最大化する$w$を求める。

\begin{equation}
\eta^2=\frac{S_B}{S_T}=\frac{S_B}{S_B+S_W}=\frac{S_B/S_W}{S_B/S_W+1}
\end{equation}

すなわち、相関比$\eta^2$の最大化は$S_B/S_W$の最大化と同等であり、$S_B/S_W$を最大化する$w$は次式で与えられる。<br>

\begin{equation}
Ww = \left ( \bar{x}^{A}-\bar{x}^{B} \right )
\end{equation}

したがって、

\begin{align}
\left(
\begin{array}{c}
w_1 \\
w_2
\end{array}
\right)
&=
\left(
\begin{array}{cc}
S_{11}^{A}+S_{11}^{B}  & S_{12}^{A}+S_{12}^{B} \\
S_{12}^{A}+S_{12}^{B} & S_{22}^{A}+S_{22}^{B}
\end{array}
\right)^{-1}
\left(
\begin{array}{c}
\bar{x}_{1}^{A}-\bar{x}_{1}^{B}\\
\bar{x}_{2}^{A}-\bar{x}_{2}^{B}
\end{array}
\right)
\end{align}


また、$w_0$は

\begin{equation}
w_0 =
-\left(
\begin{array}{cc}
w_{1} &w_{2}
\end{array}
\right)
\left(
\begin{array}{c}
\bar{x}_{1}\\
\bar{x}_{2}
\end{array}
\right)
\end{equation}

したがって、分割直線と判別関数は

\begin{align}
& x_2=-\frac{w_{1}}{w_{2}}x_1-\frac{w_{1}}{w_{2}}w_0\\
& z=w_1x_1+w_2x_2+w_0
\end{align}

一般的には、関数$f$は次のように書くことができ、これをフィッシャーの線形判別関数と呼ぶ。

\begin{align}
f &= w^\mathrm{T}x\\
&= x^\mathrm{T}w\\
&= x^\mathrm{T}W^{-1}\left ( \bar{x}^{A}-\bar{x}^{B} \right )
\end{align}




<h3> 3．　マハラノビスの距離による判別分析（2変量2群の判別の場合）</h3>

いま、2変量$x=(x_1,x_2)^\mathrm{T}$の2群$A,B$の判別を考える。

群$A$における$x$の確率分布を$N(\bar{x}^A,\Sigma^A)$、群$B$における$x$の確率分布を$N(\bar{x}^B,\Sigma^B)$とする。ただし、$\Sigma$は共分散行列である。

\begin{align}
\bar{x}^A &=
\left(
\begin{array}{c}
\bar{x}_1^A \\
\bar{x}_2^A 
\end{array}
\right)\\
\bar{x}^B &=
\left(
\begin{array}{c}
\bar{x}_1^B \\
\bar{x}_2^B 
\end{array}
\right)\\
\Sigma^A &=
\left(
\begin{array}{cc}
\sigma_{1}^{2A} & \sigma_{12}^A \\
\sigma_{12}^A & \sigma_{2}^{2A} 
\end{array}
\right) \\
\Sigma^B &=
\left(
\begin{array}{cc}
\sigma_{1}^{2B} & \sigma_{12}^B \\
\sigma_{12}^B & \sigma_{2}^{2B} 
\end{array}
\right) 
\end{align}

ここで、$x$の各群への距離をマハラノビスの距離の2乗で表せば、

\begin{align}
(D^A)^2 &= (x-\bar{x}^A)^\mathrm{T}\frac{1}{\Sigma^A}(x-\bar{x}^A) \\
&= \frac{(x-\bar{x}^A)^{2}}{\sigma^{2}} \\
(D^B)^2 &= \frac{(x-\bar{x}^B)^{2}}{\sigma^{2}}
\end{align}

ここで、マハラノビスの距離の2乗$(D^{\{A,B\}})^2$は$x$から母集団$\{A,B\}$の母平均ベクトル$\bar{x}^{\{A,B\}}$までの距離を測る量である。$\Sigma^{\{A,B\}}=I_2$（単位行列）ならば、$(D^{\{A,B\}})^2$はユークリッド距離となるが、通常は$\Sigma^{\{A,B\}} \neq I_2$なので、$(D^{\{A,B\}})^2$は分散を用いてユークリッド距離を調整した量である。こういった意味で、この距離をマハラノビスの汎距離（Mahalanobis' generalized distance）という。したがって、$(D^B)^2>(D^A)^2$ならば$x$は群$A$に、$(D^B)^2<(D^A)^2$ならば$x$は群$B$に属する。

<h4> 5.1　Pythonによるコーディング</h4>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#データの読み込み
data = pd.read_csv("data_da_F4.csv",sep=",", usecols=[1,2], skiprows=[16])
#print(data)

N2=8
N1=7
x1_unknown=4
x2_unknown=6

data=np.array(data)

G1=np.array([data[i] for i in range(0,N1)])
print("G1=\n",G1)

G2=np.array([data[i] for i in range(N1,N1+N2)])
print("G2=\n",G2)

print("\n")
print("Fisher's Linear Discriminant\n")

M=np.mean(data,axis=0)
print("M=",M)

W=np.zeros((2,2))

m1=np.mean(G1,axis=0)
m1=m1.reshape(2,1)
print("m1=",m1)
for xn in G1:
    xn=xn.reshape(2,1)
    W+=np.dot((xn-m1),(xn-m1).T)

m2=np.mean(G2,axis=0)
m2=m2.reshape(2,1)
print("m2=",m2)
for xn in G2:
    xn=xn.reshape(2,1)
    W+=np.dot((xn-m2),(xn-m2).T)

print("W=", W)
W_inv=np.linalg.inv(W)
print("W_inv=\n",W_inv)
w=np.dot(W_inv,m1-m2)
print("w=", w)
w0=-np.dot(M,w)
print("w0=",w0)


a=w[0]/w[1]
b=w0/w[1]
print("a=",a)
print("b=",b)

w=w.T
print("z=")
for xn in G1:
    xn=xn.reshape(2,1)
    z1=np.dot(w,xn)+w0
    print(z1)
for xn in G2:
    xn=xn.reshape(2,1)
    z2=np.dot(w,xn)+w0
    print(z2)


x_unknown=np.array([[x1_unknown],[x2_unknown]])
z_unknown=np.dot(w,x_unknown)+w0
print("z_unknown=",z_unknown)

x, y = G1.T
plt.plot(x, y, 'bo', ms=5, label='class1')
x, y = G2.T
plt.plot(x, y, 'ro', ms=5, label='class2')
x, y = x_unknown
plt.plot(x, y, 'go', ms=5, label='unknown')


line_x = np.linspace(0, 10, 10)
line_y = -a*line_x-b

plt.plot(line_x, line_y, "k", ms=3)
plt.xlim(-2,12)
plt.ylim(0,12)
plt.show()

print("\n")
print("Mahalanobis' Distance\n")

sigma1=np.cov(G1, rowvar=0, bias=1)
print("sigma1=", sigma1)
sigma1_inv=np.linalg.inv(sigma1)
print("sigma1_inv=", sigma1_inv)

for xn in G1:
    xn=xn.reshape(2,1)
    D1=np.dot((xn-m1).T, sigma1_inv)
    D1=np.dot(D1, (xn-m1))
    print("D1^2=", D1)
for xn in G2:
    xn=xn.reshape(2,1)
    D1=np.dot((xn-m1).T, sigma1_inv)
    D1=np.dot(D1, (xn-m1))
    print("D1^2=", D1)

print("\n")

sigma2=np.cov(G2, rowvar=0, bias=1)
print("sigma2=", sigma2)
sigma2_inv=np.linalg.inv(sigma2)
print("sigma2_inv=", sigma2_inv)

for xn in G1:
    xn=xn.reshape(2,1)
    D2=np.dot((xn-m2).T, sigma2_inv)
    D2=np.dot(D2, (xn-m2))
    print("D2^2=", D2)
for xn in G2:
    xn=xn.reshape(2,1)
    D2=np.dot((xn-m2).T, sigma2_inv)
    D2=np.dot(D2,(xn)-m2)
    print("D2^2=", D2)

print("\n")
D1=np.dot((x_unknown-m1).T, sigma1_inv)
D1=np.dot(D1, (x_unknown-m1))
D2=np.dot((x_unknown-m2).T, sigma2_inv)
D2=np.dot(D2, (x_unknown-m2))
print("D1^2, D2^2 of unknown x=", D1, D2)
```
