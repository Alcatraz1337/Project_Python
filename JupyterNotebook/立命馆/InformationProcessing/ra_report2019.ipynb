<h3> 1．　単回帰分析</h3><br>
単回帰モデルは以下の式で表される。<br>
\begin{equation}
y_i =\beta_0+\beta_1x_i+\varepsilon_i \qquad(i=1, \cdots, n; \varepsilon_i \sim \, N(0, \sigma^2))
\end{equation}

回帰式は
\begin{equation}
\hat{y} =\beta_0+\beta_1x
\end{equation}

残差は
\begin{equation}
e_i =y_i-(\beta_0+\beta_1x_i)
\end{equation}

$n$個の残差$e_i$の平方和（残差平方和）　は
\begin{equation}
S_e=\sum_{i=1}^{n}e_i^2
\end{equation}

$S_e$が最小となる切片$\beta_0$と係数$\beta_1$は
\begin{align}
&\beta_0=\bar{y}-\beta_1\bar{x}\\
&\beta_1=\frac{S_{xy}}{S_{xx}}
\end{align}

回帰による平方和は
\begin{equation}
S_R=\frac{S_{xy}^2}{S_{xx}}
\end{equation}

推定した回帰式$\hat{y}$がどの程度有効なのかを表す指標である寄与率は
\begin{equation}
R^2=1-\frac{S_e}{S_{yy}}
\end{equation}

<h3> 2．　重回帰分析</h3>
<h4>2.1 説明変数が2個の場合</h4>
重回帰モデルは以下の式で表される。

\begin{equation}
y_i =\beta_0+\beta_1x_{1i}+\beta_2x_{2i}+\varepsilon_i \qquad(i=1, \cdots, n; \varepsilon_i \sim \, N(0, \sigma^2))
\end{equation}

回帰式は
\begin{equation}
\hat{y} =\beta_0+\beta_1x_{1i}+\beta_2x_{2i}
\end{equation}

残差平方和$S_e$が最小となる切片$\beta_0$および係数$\beta_1,\beta_2$は

\begin{align}
    &\beta_0=\bar{y}-\beta_1\bar{x_1}-\beta_2\bar{x_2}\\
    &\beta_1=\frac{S_{22}S_{1y}-S_{12}S_{2y}}{S_{11}S_{22}-S_{12}^2}\\
    &\beta_2=\frac{-S_{12}S_{1y}+S_{11}S_{2y}}{S_{11}S_{22}-S_{12}^2}
\end{align}

回帰による平方和は
\begin{equation}
S_R=\beta_1S_{1y}+\beta_2S_{2y}
\end{equation}

推定した回帰式$\hat{y}$がどの程度有効なのかを表す指標である寄与率は

\begin{equation}
R^2=1-\frac{S_3}{S_{yy}}
\end{equation}

<h4>2.2 説明変数が3個の場合</h4>
重回帰モデルは以下の式で表される。

\begin{equation}
y_i =\beta_0+\beta_1x_{1i}+\beta_2x_{2i}+\beta_3x_{3i}+\varepsilon_i \qquad(i=1, \cdots, n; \varepsilon_i \sim \, N(0, \sigma^2))
\end{equation}

回帰式は

\begin{equation}
\hat{y}_i =\beta_0+\beta_1x_{1i}+\beta_2x_{2i}+\beta_3x_{3i}
\end{equation}

残差平方和$S_e$が最小となる切片は

\begin{equation}
\beta_0=\bar{y}-\beta_1\bar{x_1}-\beta_2\bar{x_2}-\beta_3\bar{x_3}
\end{equation}

$\beta_1,\,\beta_2,\,\beta_3$は次式で与えられる。

\begin{equation}
\left(
\begin{array}{c}
\beta_1\\
\beta_2\\
\beta_3
\end{array}
\right) =
\left(
\begin{array}{ccc}
S_{11}  & S_{12}   & S_{13}  \\
S_{21}  & S_{22}   & S_{23}  \\
S_{31}  & S_{32}   & S_{33} 
\end{array}
\right)^{-1}
\left(
\begin{array}{c}
S_{1y}\\
S_{2y}\\
S_{3y}\\
\end{array}
\right)
\end{equation}

回帰による平方和は

\begin{equation}
S_R=\beta_1S_{1y}+\beta_2S_{2y}+\beta_3S_{3y}
\end{equation}

寄与率は

\begin{equation}
R^2=1-\frac{S_R}{S_{yy}}=r_{y\hat{y}}^2
\end{equation}

<h3> 4．　Pythonを用いた重回帰分析</h3>

   </ul><br>
    <li>Pythonでコーディングをして、サンプルデータを重回帰分析を実行せよ。</li>

```python
import pandas as pd
import numpy as np

data=pd.read_csv("data_ra_F2.csv", sep=",")
print(data.head())

from sklearn import linear_model
clf=linear_model.LinearRegression()

data_except_y=data.drop("y",axis=1)
X=data_except_y.values
y=data['y'].values

clf.fit(X, y)

print(pd.DataFrame({"Name":data_except_y.columns,
    "Coefficients":clf.coef_}).sort_values(by='Coefficients'))
print("SEPPEN=",clf.intercept_)
print("KIYORITSU=",clf.score(X, y))
```
